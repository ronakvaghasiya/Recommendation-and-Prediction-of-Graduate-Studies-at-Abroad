```python
from tkinter import *
from tkinter import ttk
from tkinter import filedialog
from tkinter import messagebox
from tkinter import Canvas
import tkinter.font as font
from tkinter.ttk import Frame, Button, Entry, Style, Label,Combobox
import pandas as pd
```

```python
from tkinter import *
import tkinter.ttk
import pickle
import numpy as np
import tkinter as tk
model=pickle.load(open('Adchancemodel_final','rb'))

usadata = pd.read_csv("US_RANK.csv")
list_us=[]
for i in usadata.University:
    list_us.append(i)

ukdata = pd.read_csv("UK_RANK.csv")
list_uk=[]
for i in ukdata.University:
    list_uk.append(i)
    
gerdata = pd.read_csv("GERMANY_RANK.csv")
list_germany=[]
for i in gerdata.University:
    list_germany.append(i)

candata = pd.read_csv("CANADA_RANK.csv")
list_canada=[]
for i in candata.University:
    list_canada.append(i)
    
ausdata = pd.read_csv("AUSTRALIA_RANK.csv")
list_aus=[]
for i in ausdata.University:
    list_aus.append(i)
    
nzdata = pd.read_csv("NEWZEALAND_RANK.csv")
list_nz=[]
for i in nzdata.University:
    list_nz.append(i)

category = {'USA': list_us,
            'UK': list_uk,
            'GERMENY':list_germany,
           'NEW ZEALAND':list_nz,
           'AUSTRALIA':list_aus,
           'CANADA':list_canada}

def submit():
    gre=GRE.get()
    print("GRE=", gre)
    toefl=TOEFL.get()
    print("TOEFL= ", toefl)
    cgpa=float(CGPA.get())
    print("CGPA = ",cgpa)
    lorr=float(lor.get())
    print("LOR = ", lorr)
    research=r.get()
    print("Research = ", research)
    ctry = country.get()
    print("Country = ",ctry)
    uni = university.get()
    print("name of University = ",uni)
    
    if ctry == "USA":
        f1=usadata[usadata.University==uni].Rank.index[0]
    if ctry == "UK":
        f1=ukdata[ukdata.University==uni].Rank.index[0]
    if ctry == "GERMENY":
        f1=gerdata[gerdata.University==uni].Rank.index[0]
    if ctry == "NEW ZEALAND":
        f1=nzdata[nzdata.University==uni].Rank.index[0]
    if ctry == "AUSTRALIA":
        f1=ausdata[ausdata.University==uni].Rank.index[0]
    if ctry == "CANADA":
        f1=candata[candata.University==uni].Rank.index[0]
        
    print(f1)
    features=np.array([[gre,toefl,cgpa,lorr,research,f1]])
    chance_per=model.predict(features)
    cp=round(chance_per[0][0]*100)
    print("Chance of Admission Percentage = ",cp)
    message = "Your chances of admission is "+str(cp)
    messagebox.showinfo("Chance of Admission",message)
    
        
'''       b=int(app.variable_toefl.get())
        c=float(self.variable_lor.get())
        d=float(self.variable_cgpa.get())
        e=int(self.variable_research.get())
        f=int(f1)
        print("GRE=",a," Toefl =",b," LOR = ",c," CGPA=",d," research=",e," rank = ",f)
        features=np.array([[a,b,c,d,e,f]])
        chance=model.predict(features)
        print(chance)
        message = "Your chances of admission in ",self.AccountCombo.get()," is :", str(chance)
        messagebox.showinfo("Chance of Admission",message)'''
        
def getUpdateData(event):
    university['values'] = category[country.get()]

top = Tk()
top.title('Show Acceptance Estimates')
top.geometry("800x600")

GRE = IntVar()
TOEFL = IntVar()
CGPA = DoubleVar()

top.columnconfigure(0, pad=3)
top.columnconfigure(1, pad=3)
top.columnconfigure(2, pad=3)
top.columnconfigure(3, pad=3)

top.rowconfigure(0, pad=3)
top.rowconfigure(1, pad=3)
top.rowconfigure(2, pad=3)
top.rowconfigure(3, pad=3)
top.rowconfigure(4, pad=3)

grescore = Label(top, text = "GRE")
grescore.grid(row=0, column=0, sticky = W+E, padx=50, pady=30)
gre = Entry(top,textvariable = GRE)
gre.grid(row=0, column=1, sticky = W+E)

toeflscore=Label(top, text = "TOEFL")
toeflscore.grid(row=1, column=0, sticky = W+E, padx=50)
toefl = Entry(top,textvariable = TOEFL)
toefl.grid(row=1, column=1, sticky=W+E)

cgpascore = Label(top, text = "CGPA")
cgpascore.grid(row=2, column=0, sticky = W+E, padx=50, pady=30)
cgpa = Entry(top,textvariable=CGPA)
cgpa.grid(row=2, column=1, sticky=W+E)

lorrate = Label(top, text = "LOR")
lorrate.grid(row=3, column=0, sticky = W+E, padx=50)
lor = ttk.Combobox(values = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])
lor.grid(row=3,column=1, sticky=W+E)

research = Label(top, text = "RESEARCH")
research.grid(row=4, column=0, sticky = W+E, padx=50,pady = 30)
r = tk.IntVar()
r0 = tk.Radiobutton(text='0', variable=r)
r0.config(indicatoron=0, bd=4, width=5, value='0')
r0.grid(row=4, column=1, sticky=W)
r1 = tk.Radiobutton(text='1', variable=r)
r1.config(indicatoron=0, bd=4, width=5, value='1')
r1.grid(row=4, column=1, sticky=E)
        
ctry = Label(text = 'Country:').grid(row = 5,column = 0, sticky = W+E, padx = 50) 
country = ttk.Combobox(values = list(category.keys()))
country.bind('<<ComboboxSelected>>', getUpdateData)
country.grid(row = 5,column = 1,sticky=W+E)

uni = Label(text = 'University:').grid(row = 6,column = 0, sticky = W+E, padx = 50, pady=30)
university = ttk.Combobox( width = 35)
university.grid(row = 6,column=1)


sub = Button(top, text="Submit",command=submit)
sub.grid(row=7,column=1)

if __name__ == '__main__':
    top.mainloop()
```
